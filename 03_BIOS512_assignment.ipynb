Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.height = 6, repr.plot.width = 8, repr.plot.res = 250)
p = ggplot()
p = p + geom_bar(data = penguins, mapping = aes(x = body_mass_g))
p = p + scale_x_continuous(labels = scales::comma_format(big.mark = ',')) + labs(x = 'Body Mass (g)', y = 'Penguin Count')
p = p + facet_wrap(~year, nrow = 3)
p = p + theme(panel.background = element_rect(fill = "white", colour = "grey50"), panel.border = element_rect(linetype = "dashed", fill = NA))

p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.height = 6, repr.plot.width = 8, repr.plot.res = 250)
p = ggplot()
p = p + geom_jitter(data = penguins, mapping = aes(x = body_mass_g, y = sex), height = 0.25)
p = p + labs(x = 'Body Mass (g)')
p = p + facet_wrap(~species, nrow = 3)
p = p + theme(panel.background = element_rect(fill = "#4B9CD3", colour = "white"), panel.border = element_rect(linetype = "dashed", fill = NA))

p
```
