```r
library('tidyverse')
```

```r
url <- 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv'

turbine <- read_csv(url, col_type = cols())
head(turbine)
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
turbine |>
    group_by(province_territory) |>
    summarize(turbine_count = n()) |>
    arrange(desc(turbine_count))
```

```r
turbine_plot = turbine |>
    group_by(province_territory) |>
    mutate(turbine_count = n()) |>
    ungroup() |>
    arrange(turbine_count) |>
    mutate(province_territory = fct_inorder(province_territory))

# head(turbine_plot)
```

```r
options(repr.plot.height = 3, repr.plot.width = 10, repr.plot.res = 250)
p = ggplot(turbine_plot, aes(y = province_territory)) +
    geom_bar()
p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
turbine |>
    mutate(manufacturer = manufacturer |> fct_lump_min(10)) |>
    group_by(manufacturer) |>
    summarize(manu_count = n()) |>
    arrange(desc(manu_count))
```

```r
manu_plot = turbine |>
    mutate(manufacturer = manufacturer |> fct_lump_min(10)) |>
    group_by(manufacturer) |>
    mutate(manu_count = n()) |>
    ungroup() |>
    arrange(manu_count) |>
    mutate(manufacturer = fct_inorder(manufacturer) |> 
#            fct_rev() if wanted in reverse order
           fct_relevel('Other'))

levels(manu_plot$manufacturer)
```

```r
options(repr.plot.height = 3, repr.plot.width = 10, repr.plot.res = 250)
p = ggplot() +
    geom_bar(manu_plot, mapping = aes(y = manufacturer))
p
```
